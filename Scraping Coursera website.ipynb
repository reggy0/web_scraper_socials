## Gathering Data <a id= gathering></a>  [Top](#top)
> ### I Gatherd Data from www.courera.org by webscraping. I used Selenium , Requests , Beautifulsoup and pandas python libirties to extract data and store it in structered way.

```python
# importing required libirtied
from selenium import webdriver
from selenium.webdriver.chrome.options import Options as ChromeOptions
import time
from bs4 import BeautifulSoup
import pandas as pd
import json
import requests
from selenium.webdriver.support.select import Select
import re
```

```python
# creating lists for required data for the first dataframe
Title = []
Partner = []
Type = []
Level = []
link = []
students = []
Rate = []
```

```python
page = 1

# Add additional Options to the webdriver
chrome_options = ChromeOptions()
# add the argument and make the browser Headless.
chrome_options.add_argument("--headless")
# intiate the driver
driver = webdriver.Chrome(options = chrome_options)

for i in range (1):
    url_page= 'https://www.coursera.org/search?query=Data%20Analysis&page={}&index=prod_all_products_term_optimization&topic=Data%20Science&allLanguages=English&skills=Data%20Analysis'.format(page)
    # get the web page   
    driver.get(url_page)
    time.sleep(5)
    # take the page source and pass it to BeautifulSoup
    pageSource = driver.page_source
    soup = BeautifulSoup(pageSource, 'lxml')
    
    courses = soup.find("main", {"id": "main"})
#except Exception:
    #driver.close()
    #loadpage()

    if len(courses.find('ul' , class_="ais-InfiniteHits-list").find_all('h2' , class_ ='color-primary-text card-title headline-1-text' )) == 0:
        course_name = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('h2' , {'class' : 'cds-1 card-title css-iyr9nj cds-3'} )
    else:
        course_name = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('h2' , {'class' : 'color-primary-text card-title headline-1-text'} )

    course_partner = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('span' , class_='cds-1 css-mx49ok cds-3')
    
    if len(courses.find('ul' , class_="ais-InfiniteHits-list").find_all('div' , class_='product-type-pill-wrapper')) == 0:
        course_type = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('div' , class_='rc-ProductTypePill css-155u7q3')
    else:
        course_type = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('div' , class_='product-type-pill-wrapper')
        
        
        
    if len(courses.find('ul' , class_="ais-InfiniteHits-list").find_all('div' , 'product-difficulty'))  == 0:
        
        course_level = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('span' , 'cds-1 difficulty css-lqm5si cds-3')
        
    else:    
        course_level = courses.find('ul' , class_="ais-InfiniteHits-list").find_all('div' , 'product-difficulty')
    
    
    course_stud = courses.find_all('span' , class_ = 'enrollment-number')




    links = soup.find('script' , type="application/ld+json")
    links = str(links)
    links = links.replace('<script type="application/ld+json">' , "").replace('</script>' , "")[11:]
    data = json.loads(links)

    course_name = [ x.text for x in course_name]
    course_partner = [ x.text for x in course_partner]
    course_type = [ x.text for x in course_type]
    course_level = [x.text for x in course_level]
    
    course_rate = []
    for x in courses.find_all('li' , class_ = 'ais-InfiniteHits-item'):
        if x.find('span' , {'class' : 'ratings-text' , 'aria-hidden' : "true"} ) == None:
            course_rate.append('None')
        else:
            course_rate.append(x.find('span' , class_ = 'ratings-text').text)

    course_stud = []       
    for x in courses.find_all('li' , class_ = 'ais-InfiniteHits-item'):
        if x.find('span' , class_ = 'enrollment-number' ) == None:
            course_stud.append('None')
        else:
            course_stud.append(x.find('span' , class_ = 'enrollment-number').text)
#course_stud = soup.find_all('span' , class_ = 'enrollment-number') 


    #aria-hidden="true"      

    for i in course_name:
        Title.append(i)       
    for i in course_partner:
        Partner.append(i)
    for i in course_type:
        Type.append(i)
    for i in course_level:
        Level.append(i)
    for i in data['itemListElement']:
        url = i['url']
        link.append(url)
    for i in course_rate:
        Rate.append(i)
    for i in course_stud:
        students.append(i)
                
    page +=1
    
driver.close()
```

```python
 
for i in link:
    try:
        path = r'C:\chromedriver\chromedriver.exe'
        driver = webdriver.Chrome(executable_path = path)
        driver.get(i)
        time.sleep(10)
        pageSource = driver.page_source
        soup = BeautifulSoup(pageSource, 'html.parser')
        #html = requests.get('https://www.coursera.org/learn/introduction-to-data-analytics').text
        #time.sleep(5)
        # extracting data from html
        #soup = BeautifulSoup(html , 'html.parser')
        
        # skills will be a list. we want thtis list te have one 'none' item if it is empty or the right items 
        # if they exist.
        if soup.find('div' , class_ = '_t6niqc3') == None:
            pre_skills = ['None']
            Skills.append('None')
        else:
            pre_skills = soup.find('div' , class_ = '_t6niqc3').find_all('span' , class_ = '_1q9sh65')
            for skill in pre_skills:
                Skills.append(skill.text) 
       # eacth of title, rating, and students will be  single item. we whant it to be 'none if it does not exsit and the right
        # item If it exists 
        
        if soup.find('span' , {'data-test' : "number-star-rating" }) == None:         
            pre_rating = 'None'
        else:
            pre_rating = soup.find('span' , {'data-test' : "number-star-rating" }).text
            

        if soup.find('div' , {'data-test' : "banner-title-container"} ) == None:
            pre_title = 'None'
        else:
            pre_title = soup.find('div' , {'data-test' : "banner-title-container"} ).find('h1').text
            

        if soup.find('div' , class_ = '_1fpiay2') == None:
            
            pre_students = 'None'
        else:     
            pre_students = soup.find('div' , class_ = '_1fpiay2').text
            

        # we want to convert title, ranint, and studens into lists with the same skills number of items so we can convert all lists
        # into dataframe later

        for i in range(len(pre_skills)):
            Title2.append(pre_title)
        for i in range(len(pre_skills)):
            Students.append(pre_students)
        for i in range(len(pre_skills)):
            Rating.append(pre_rating)
        driver.close()
    except ConnectionError:
        print('Connection Error in link: {}'.format(i))
        time.sleep(5)
        continue
```
